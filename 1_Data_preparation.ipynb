```python
outcomes_list = ['STATUS OS','OS','STATUS PFS','PFS','DCR','OS-6','OS-18','OS-24','PFS-3','PFS-9']
```

```python
# Separate external validation set

mask_external = df['Center'].astype(str).str.strip() == "Padova"
df_internal = df[~mask_external]
print('Number of patients in the internal set:', len(df_internal))
df_extval = df[mask_external]
print('Number of patients in the external validation set (from Padova):', len(df_extval))

X_extval = df_extval.drop(columns=outcomes_list)
y_extval = df_extval[outcomes_list]
```

```python
# EXPLORE MISSINGNESS IN THE WHOLE DATABASE

# Store the names of the centers in a list
centers_list = df["Center"].dropna().unique().tolist()
# Center column only
df_center = df[["Center"]].copy()


# Function for the blue-yellow heatmap:
# it is useful to understand if some features are preferably missing in specific centers

from matplotlib.colors import ListedColormap

def create_heatmap(df_index: pd.DataFrame, df: pd.DataFrame, title, color_map="YlGnBu", colors=None, xlabel='', ylabel='') -> plt:
    if not colors is None:
        color_map = ListedColormap([colors[0], colors[1]])

    # Sort df_index by 'Center'
    df_index.sort_values(by='Center', inplace=True)
    # Reindex data to match the sorted order of df_index
    data = df.copy().reindex(df_index.index)
    groups = list(df_index['Center'].apply(lambda x: next((c for c in centers_list if x.startswith(c)), None)))
    unique_labels = sorted(set(groups), key=groups.index)
    positions = [groups.index(label) for label in unique_labels]

    sns.set_style("whitegrid")
    sns.set_theme(style="ticks")

    plt.figure(figsize=(15, 20))
    plt.title(title, fontsize=20)

    ax = sns.heatmap(data.isnull(), cmap=color_map, cbar=False, yticklabels=False)
    for p in positions:
        plt.axhline(p, color='black', linewidth=0.5)
    plt.xticks(np.arange(0.5, len(data.columns) + 0.5), labels=np.array(data.columns), fontsize=10, rotation=90)
    plt.yticks(np.arange(0.5, len(groups) + 0.5), labels=np.array(groups), fontsize=10)
    plt.gca().set_yticks(positions)
    plt.gca().set_yticklabels(unique_labels)
    ax.set_xticklabels(data.columns)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    sns.despine()

    return plt


# Blue-yellow heatmap of missing values in the training set
color1 = "#49659B"# (0 / 255, 102 / 255, 204 / 255)
color2 = (255 / 255, 255 / 255, 0 / 255)

create_heatmap(df_center, df, "", colors=[color1, color2], xlabel='', ylabel='')
```

```python
# More than 30% of LDH is missing, this feature is excluded

X_train = X_train.drop(['LDH at baseline'], axis=1)

X_test = X_test.drop(['LDH at baseline'], axis=1)
X_extval = X_extval.drop(['LDH at baseline'], axis=1)
```

# Imputation of missing values

```python
# One-hot-encoding for 'Center'
from sklearn.preprocessing import OneHotEncoder

# Create dummy variables
encoder_center = OneHotEncoder()
encoder_center.fit(X_train[['Center']])
center_encoded_train = encoder_center.transform(X_train[['Center']])
# DataFrame with the dummy variables
transformed_center_train = pd.DataFrame(center_encoded_train.toarray(),
                                        columns=encoder_center.get_feature_names_out(),
                                        index=X_train.index)
# Put dummy variables into the main DataFrame
X_train = pd.concat([X_train, transformed_center_train], axis=1)
# Dropping old categorical column 'Center'
X_train.drop(['Center'], axis=1, inplace=True)

# Do the same for test (extval has NO missing, so imputation is not done!)

# TEST
center_encoded_test = encoder_center.transform(X_test[['Center']])
# DataFrame with the dummy variables
transformed_center_test = pd.DataFrame(center_encoded_test.toarray(),
                                        columns=encoder_center.get_feature_names_out(),
                                        index=X_test.index)
# Put dummy variables into the main DataFrame
X_test = pd.concat([X_test, transformed_center_test], axis=1)
# Dropping old categorical column 'Center'
X_test.drop(['Center'], axis=1, inplace=True)
```

```python
# IMPUTATION OF MISSING VALUES

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
# Initialize the IterativeImputer
imputer = IterativeImputer(random_state=63)
# Fit the imputer on the training set
imputer.fit(X_train) # column 'Center' is used for imputation!!!
# Transform the three datasets
X_train_imputed = pd.DataFrame(imputer.transform(X_train), columns=X_train.columns)
X_test_imputed = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)

X_train_imputed
```

```python
# Remove columns starting with 'Center_'
cols_to_drop = [c for c in X_train.columns if c.startswith('Center')]
X_train_imputed.drop(columns=cols_to_drop, inplace=True)
X_train_imputed.head()

X_test_imputed.drop(columns=cols_to_drop, inplace=True)
X_extval.drop(columns='Center', inplace=True)
```

```python
# Round the imputed values

for var in X_train_imputed.columns:
    if var != 'Hb at baseline':
        X_train_imputed[var] = round(X_train_imputed[var]).astype('int64')
        X_test_imputed[var] = round(X_test_imputed[var]).astype('int64')
    else:
        X_train_imputed[var] = round(X_train_imputed[var], 2)
        X_test_imputed[var] = round(X_test_imputed[var], 2)
```

# Build NLR, PLR, SII

```python
for set in (X_train_imputed, X_test_imputed, X_extval):
    set['NLR at baseline'] = round(set['Neutrophils at baseline'] / set['Lymphocytes at baseline'], 2)
    set['PLR at baseline'] = round(set['Platelets at baseline'] / set['Lymphocytes at baseline'], 2)
    set['SII at baseline'] = round(set['Platelets at baseline'] * set['NLR at baseline'], 2)
```

## Logarithmic transformation

```python
# Perform log transformation on all numerical variables
import math

numerical_log = []

for var in numerical_initial:
    # Append the variable name and its log-transformed version to the list
    numerical_log.append(var)
    numerical_log.append(f'log_{var}')
    # Apply log transformation to the variable and create a new column
    X_train_imputed[f'log_{var}'] = X_train_imputed[var].apply(lambda y: math.log(y + 1))
```

```python
# Plot the histograms of the original and logarithmic numerical features

# Set the figure size
plt.figure(figsize=(15, 20))
# Adjust the font size
plt.rc('font', size=20)
sns.set_style("whitegrid")
# Loop over each numerical variable
for i, var in enumerate(numerical_log):
    plt.subplot(7, 4, i+1)  # Adjust grid parameters to fit your number of variables
    sns.histplot(x=var, data=X_train_imputed, bins=30)
    plt.tight_layout()
#plt.show()
```

```python
# Statistical test for the normality of the numerical features

from scipy.stats import shapiro

for i,var in enumerate(numerical_log):
    stat, p = shapiro(X_train_imputed[var])
    #print(f'Statistics={stat}, p={p}')
    if p > 0.05:
        print('{s} looks Gaussian (fail to reject H0)'.format(s=var))
        if i % 2 != 0:
            print('---')
    else:
        #print('{s} does NOT look Gaussian (reject H0)'.format(s=var))
        if i % 2 != 0:
            print('---')
```

```python
# SAVE separately for each outcome - REMOVING PATIENTS missing the outcome

outs = ['OS','PFS','DCR','OS-6','OS-18','OS-24','PFS-3','PFS-9']

# Initialize the dictionaries
X_train_final = {}
X_test_final = {}
X_extval_final = {}
y_train_final = {}
y_test_final = {}
y_extval_final = {}
X_train_expl_final = {}
X_test_expl_final = {}
X_extval_expl_final = {}

for i, outcome in enumerate(outs):
    for set in ('train', 'test', 'extval'):
    
        if outcome == 'OS' or outcome == 'PFS': # SURVIVAL ANALYSIS
            task = 'survival_analysis'
            
            if outcome == 'OS':
                status = 'STATUS OS'
            else:
                status = 'STATUS PFS'
            
            # TRAIN SET
            X_train_final[i] = X_train_imputed.loc[y_train.dropna(subset=[status, outcome]).index]
            y_train_final[i] = y_train.loc[y_train.dropna(subset=[status, outcome]).index]
            y_train_final[i] = y_train_final[i][[status, outcome]]
            X_train_expl_final[i] = X_train_expl.loc[y_train.dropna(subset=[status, outcome]).index]

            # TEST SET
            X_test_final[i] = X_test_imputed.loc[y_test.dropna(subset=[status, outcome]).index]
            y_test_final[i] = y_test.loc[y_test.dropna(subset=[status, outcome]).index]
            y_test_final[i] = y_test_final[i][[status, outcome]]
            X_test_expl_final[i] = X_test_expl.loc[y_test.dropna(subset=[status, outcome]).index]

            # EXTVAL SET
            X_extval_final[i] = X_extval.loc[y_extval.dropna(subset=[status, outcome]).index]
            y_extval_final[i] = y_extval.loc[y_extval.dropna(subset=[status, outcome]).index]
            y_extval_final[i] = y_extval_final[i][[status, outcome]]
            X_extval_expl_final[i] = X_extval_expl.loc[y_extval.dropna(subset=[status, outcome]).index]
        
        else: # CLASSIFICATION
            task = 'classification'

            # TRAIN SET
            X_train_final[i] = X_train_imputed.loc[y_train.dropna(subset=[outcome]).index]
            y_train_final[i] = y_train.loc[y_train.dropna(subset=[outcome]).index]
            y_train_final[i] = y_train_final[i][[outcome]]
            X_train_expl_final[i] = X_train_expl.loc[y_train.dropna(subset=[outcome]).index]
            
            # TEST SET
            X_test_final[i] = X_test_imputed.loc[y_test.dropna(subset=[outcome]).index]
            y_test_final[i] = y_test.loc[y_test.dropna(subset=[outcome]).index]
            y_test_final[i] = y_test_final[i][[outcome]]
            X_test_expl_final[i] = X_test_expl.loc[y_test.dropna(subset=[outcome]).index]

            # EXTVAL SET
            X_extval_final[i] = X_extval.loc[y_extval.dropna(subset=[outcome]).index]
            y_extval_final[i] = y_extval.loc[y_extval.dropna(subset=[outcome]).index]
            y_extval_final[i] = y_extval_final[i][[outcome]]
            X_extval_expl_final[i] = X_extval_expl.loc[y_extval.dropna(subset=[outcome]).index]

        # Save the data
        if save_output:

            X_train_final[i].to_excel('{t}/{o}/Data/X_train.xlsx'.format(t=task, o=outcome))
            X_test_final[i].to_excel('{t}/{o}/Data/X_test.xlsx'.format(t=task, o=outcome))
            X_extval_final[i].to_excel('{t}/{o}/Data/X_extval.xlsx'.format(t=task, o=outcome))

            y_train_final[i].to_excel('{t}/{o}/Data/y_train.xlsx'.format(t=task, o=outcome))
            y_test_final[i].to_excel('{t}/{o}/Data/y_test.xlsx'.format(t=task, o=outcome))
            y_extval_final[i].to_excel('{t}/{o}/Data/y_extval.xlsx'.format(t=task, o=outcome))

            X_train_expl_final[i].to_excel('{t}/{o}/Data/X_train_expl.xlsx'.format(t=task, o=outcome))
            X_test_expl_final[i].to_excel('{t}/{o}/Data/X_test_expl.xlsx'.format(t=task, o=outcome))
            X_extval_expl_final[i].to_excel('{t}/{o}/Data/X_extval_expl.xlsx'.format(t=task, o=outcome))
```
